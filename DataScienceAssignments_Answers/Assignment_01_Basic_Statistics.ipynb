# Standardization of Numerical Variables

Concept of Standardization:
Standardization (or Z-score normalization) is a technique used to scale numerical variables such that they have a mean (Î¼) of 0 and a standard deviation (Ïƒ) of
1. It helps in situations where variables have different units or scales, allowing them to be comparable and improving the performance of machine learning models that are sensitive to the magnitude of features (e.g., k-nearest neighbors, logistic regression).

The formula for Z-score normalization is:

z= Ïƒ/xâˆ’Î¼

Where:


z = the standardized value (Z-score)
x = the original data point
Î¼ = the mean of the data
Ïƒ = the standard deviation of the data

```python

# before standardization
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
sns.histplot(df[numerical_columns[0]], kde=True)
plt.title(f'Before Standardization: {numerical_columns[0]}')

plt.subplot(2, 3, 2)
sns.histplot(df[numerical_columns[1]], kde=True)
plt.title(f'Before Standardization: {numerical_columns[1]}')

plt.subplot(2, 3, 3)
sns.histplot(df[numerical_columns[2]], kde=True)
plt.title(f'Before Standardization: {numerical_columns[2]}')

plt.subplot(2, 3, 4)
sns.histplot(df[numerical_columns[3]], kde=True)
plt.title(f'Before Standardization: {numerical_columns[3]}')

plt.subplot(2, 3, 5)
sns.histplot(df[numerical_columns[4]], kde=True)
plt.title(f'Before Standardization: {numerical_columns[4]}')

plt.subplot(2, 3, 6)
sns.histplot(df[numerical_columns[5]], kde=True)
plt.title(f'Before Standardization: {numerical_columns[5]}')

plt.tight_layout()
plt.show()

# Standardization: z = (x - mean) / std
df_standardized = df.copy()
for column in numerical_columns:
    mean = df[column].mean()
    std = df[column].std()
    df_standardized[column] = (df[column] - mean) / std

# after standardization
plt.figure(figsize=(15, 10))

# after standardization
plt.subplot(2, 3, 1)
sns.histplot(df_standardized[numerical_columns[0]], kde=True)
plt.title(f'After Standardization: {numerical_columns[0]}')

plt.subplot(2, 3, 2)
sns.histplot(df_standardized[numerical_columns[1]], kde=True)
plt.title(f'After Standardization: {numerical_columns[1]}')

plt.subplot(2, 3, 3)
sns.histplot(df_standardized[numerical_columns[2]], kde=True)
plt.title(f'After Standardization: {numerical_columns[2]}')

plt.subplot(2, 3, 4)
sns.histplot(df_standardized[numerical_columns[3]], kde=True)
plt.title(f'After Standardization: {numerical_columns[3]}')

plt.subplot(2, 3, 5)
sns.histplot(df_standardized[numerical_columns[4]], kde=True)
plt.title(f'After Standardization: {numerical_columns[4]}')

plt.subplot(2, 3, 6)
sns.histplot(df_standardized[numerical_columns[5]], kde=True)
plt.title(f'After Standardization: {numerical_columns[5]}')

plt.tight_layout()
plt.show()
```

##Inferences from Before and After Comparison:
Before Standardization:
Numerical columns likely have different ranges, means, and standard deviations.


After Standardization:
The data should be centered around zero, with a standard deviation of 1. This ensures uniformity, making the variables more suitable for models that rely on distance calculations or gradient-based optimization.

* Data preprocessing is a crucial step in data analysis and machine learning, as it enables accurate and reliable modeling.
* Standardization and one-hot encoding are two essential preprocessing techniques.
* Standardization scales data to a common range to prevent features with large ranges from dominating others.
* One-hot encoding converts categorical variables into binary vectors, allowing them to be processed by models.
* These techniques help prevent data leakage, improve model interpretability, and enable more efficient computation.
* By applying standardization and one-hot encoding, you can ensure that your data is clean, consistent, and ready for analysis and modeling.
